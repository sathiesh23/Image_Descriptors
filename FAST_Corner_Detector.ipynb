### **FAST:**
*   FAST is **used to detect corners** in images and it is most applicable to real-time applications or resource constrained devices.
*   FAST keypoint detector is that, **for a pixel to be considered a “corner”, there must be atleast n contiguous pixels along a circular perimeter with radius r that are all either brighter or darker than the center pixel by a threshold t.**

```python
# import the necessary packages
from __future__ import print_function
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
# load the image and convert it to grayscale
# Load the image from Google Drive
image_path = '/content/drive/MyDrive/MIT/Work/Machine_Learning/Google_Colab/Machine_Learning/Image_Descriptors/FAST_KeyPoint_Detection/Chess_board.jpg'
image = cv2.imread(image_path)
orig = image.copy()
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# Display the image
plt.imshow(orig)
plt.show()
```

```python
# detect FAST keypoints in the image
detector = cv2.FastFeatureDetector_create()
kps = detector.detect(gray)
print("# of keypoints: {}".format(len(kps)))
```

```python
# loop over the keypoints and draw them
for kp in kps:
	r = int(0.5 * kp.size)
	(x, y) = np.intp(kp.pt)
	cv2.circle(image, (x, y), r, (0, 255, 255), 2)

plt.imshow(np.hstack([orig, image]))
plt.show()
```
